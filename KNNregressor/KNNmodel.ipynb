## add mean values in missing columns

```python
#datacleaning
df['total_bedrooms']=df['total_bedrooms'].replace("",np.nan)
df['total_bedrooms']=df['total_bedrooms'].astype(float)
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].mean())




```

```python
print(df.isnull().sum())

```

# CHECK OUTLIERS

### drop outliers using IQR interquartile range

```python
import numpy as np

# Checking for outliers in my target 'median_house_value'
column = 'median_house_value'

# Calculating Q1, Q3, and IQR
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1

# Lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Print outlier details
print("Count of Outliers:", len(outliers))

if len(outliers) > 0:
    print("\nIndices of Outliers:", outliers.index.tolist()) 
    print("\nValues of Outliers:")
    print(outliers[column].values)
else:
    print("\nNo outliers found.")

# Remove outliers using drop (if any)
df_cleaned = df.drop(outliers.index) if len(outliers) > 0 else df

# Print the cleaned DataFrame
print("\nData after Removing Outliers:")
print(df_cleaned)
```

```python
print("\nSummary of Cleaned Data:")
print(df_cleaned.describe())
```

## encode string values

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_cleaned['ocean_proximity'] = le.fit_transform(df_cleaned['ocean_proximity'])
```

```python
#to know which numeric value got assigned to the category
#le.classes_>>contain data in sorted order
for index, category in enumerate(le.classes_):
    print(f"{category} -> {index}")
```

# prepare dataset

```python
x = df_cleaned.iloc[:,0:9].values
y = df_cleaned.iloc[:,9].values
x.shape
```

## split dataset

```python
from sklearn.model_selection import train_test_split



""" X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0) """
X_train, X_temp, Y_train, Y_temp = train_test_split(x, y, test_size=0.2, random_state=0)

# Second split: Temporary (20%) into Validation (50%) and Test (50%)
X_eval, X_test, Y_eval, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=0)

# Check the shapes
print("X_train shape:", X_train.shape)
print("X_val shape:", X_eval.shape)
print("X_test shape:", X_test.shape)

```

## standard scalar to avoid bias

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_eval=sc.transform(X_eval)
```

```python
print(X_train)
```

```python
#check validity of standarlization by checking mean and standard deviation

mean = np.mean(X_train, axis=0)  # Column-wise mean
std = np.std(X_train, axis=0) 
print("Mean for:", mean)#should be around 0
print("Standard Deviation:", std)#should be 1
```

# train the model using KNN regressor model

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_absolute_error as mae
#inital value =2
k_initial=2
initial_model = KNeighborsRegressor(n_neighbors=k_initial,metric='minkowski',p=1)
initial_model.fit(X_train, Y_train)
validation_pred=initial_model.predict(X_eval)

val_mse = mse(Y_eval,validation_pred)
val_r2 = r2(Y_eval, validation_pred)

print("val mse:",val_mse)
print("validation_r2:",val_r2)

best_mse=val_mse
best_r2=val_r2
best_k=k_initial

```

```python
#train using different kneigbours to see which one give better performance

for k in range(3,30):
    model=KNeighborsRegressor(n_neighbors=k,metric="minkowski",p=1)
    model.fit(X_train,Y_train)
    val_pred=model.predict(X_eval)
    val_r2=r2(Y_eval,val_pred)
    val_mse=mse(Y_eval,val_pred)
    if(val_r2>best_r2 and val_mse<best_mse):
        best_mse=val_mse
        best_r2=val_r2
        best_k=k
print("best k and r2 and mse respectively: ",best_k,best_r2,best_mse)
```

```python
#after that we chooose the best model
best_model=KNeighborsRegressor(n_neighbors=best_k,metric="minkowski",p=1)
best_model.fit(X_train,Y_train)
```

## predict

```python


Y_pred = best_model.predict(X_test)
```

## evaluate 

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_absolute_error as mae

#mse=1/n(ytrue-ypredict)^2>>
#mae=1/n(ytrue-ypredicted)
#r2 1>>perfect as ito moves towards 1 it becomes better -ve>>worst case scenario
test_mse = mse(Y_test, Y_pred)

test_r2 = r2(Y_test, Y_pred)
test_mae= mae(Y_test,Y_pred)

print('Mean Squared Error:',test_mse)
print('R-squared:',test_r2)
print('Mean absolute Error:',test_mae)


```
